```python
people = {
    "first" : ["Jacob","Jane","John"],
    "last" : ["Xi","Doe","Wick"],
    "email" : ["jacobxi@outlook.com","JaneDone@outlook.com", "JohnWick@outlook.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == 'Doe'
```

```python
filt = (df['last'] == 'Doe')
```

```python
df[filt]
```

```python
df[df['last'] == 'Doe']
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
filt = (df['last'] == 'Doe') & (df['first'] == 'Jane')
```

```python
df.loc[filt, 'email']
```

```python
filt = (df['last'] == 'Doe') | (df['first'] == 'John')
```

```python
df.loc[filt, 'email']
```

```python
df.loc[~filt, 'email']
```
